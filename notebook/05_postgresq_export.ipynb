```python
import pandas as pd
import numpy as np

from sqlalchemy import create_engine
from sqlalchemy import BigInteger, DateTime, Text, Integer, Boolean, Numeric
from urllib.parse import quote_plus
from dotenv import load_dotenv
import os
from IPython.display import display
```

```python
orders = pd.read_csv(
    "../data/cleaned/cleaned_orders.csv",
    parse_dates=["created_at"]
)

order_items = pd.read_csv(
    "../data/cleaned/cleaned_order_items.csv",
    parse_dates=["created_at"]
)

order_item_refunds = pd.read_csv(
    "../data/cleaned/cleaned_order_item_refunds.csv",
    parse_dates=["created_at"]
)

products = pd.read_csv(
    "../data/cleaned/cleaned_products.csv",
    parse_dates=["created_at"]
)

website_sessions = pd.read_csv(
    "../data/cleaned/cleaned_website_sessions.csv",
    parse_dates=["created_at"]
)

website_pageviews = pd.read_csv(
    "../data/cleaned/cleaned_website_pageviews.csv",
    parse_dates=["created_at"]
)

```

```python
load_dotenv()
DB_USER = os.getenv("DB_USER")
DB_PASSWORD = quote_plus(os.getenv("DB_PASSWORD"))
DB_HOST = os.getenv("DB_HOST")
DB_PORT = os.getenv("DB_PORT")
DB_NAME = os.getenv("DB_NAME")

DATABASE_URL = (
    f"postgresql://{DB_USER}:{DB_PASSWORD}"
    f"@{DB_HOST}:{DB_PORT}/{DB_NAME}"
)

engine = create_engine(DATABASE_URL)
```

```python
products.to_sql(
    name="products",
    con=engine,
    if_exists="append",
    index=False,
    dtype={
        "product_id": BigInteger(),
        "created_at": DateTime(),
        "product_name": Text(),
        "year": Integer(),
        "month": Integer(),
        "day": Integer(),
        "hour": Integer(),
        "day_name": Text(),
    }
)

website_sessions.to_sql(
    name="website_sessions",
    con=engine,
    if_exists="append",
    index=False,
    dtype={
        "website_session_id": BigInteger(),
        "created_at": DateTime(),
        "user_id": BigInteger(),
        "is_repeat_session": Boolean(),
        "utm_source": Text(),
        "utm_campaign": Text(),
        "utm_content": Text(),
        "device_type": Text(),
        "http_referer": Text(),
        "year": Integer(),
        "month": Integer(),
        "day": Integer(),
        "day_name": Text(),
    }
)

website_pageviews.to_sql(
    name="website_pageviews",
    con=engine,
    if_exists="append",
    index=False,
    dtype={
        "website_pageview_id": BigInteger(),
        "created_at": DateTime(),
        "website_session_id": BigInteger(),
        "pageview_url": Text(),
        "year": Integer(),
        "month": Integer(),
        "day": Integer(),
        "day_name": Text(),
    }
)

orders.to_sql(
    name="orders",
    con=engine,
    if_exists="append",
    index=False,
    dtype={
        "order_id": BigInteger(),
        "created_at": DateTime(),
        "website_session_id": BigInteger(),
        "user_id": BigInteger(),
        "primary_product_id": BigInteger(),
        "items_purchased": Integer(),
        "price_usd": Numeric(10, 2),
        "cogs_usd": Numeric(10, 2),
        "year": Integer(),
        "month": Integer(),
        "day": Integer(),
        "day_name": Text(),
    }
)

order_items.to_sql(
    name="order_items",
    con=engine,
    if_exists="append",
    index=False,
    dtype={
        "order_item_id": BigInteger(),
        "created_at": DateTime(),
        "order_id": BigInteger(),
        "product_id": BigInteger(),
        "is_primary_item": Boolean(),
        "price_usd": Numeric(10, 2),
        "cogs_usd": Numeric(10, 2),
        "year": Integer(),
        "month": Integer(),
        "day": Integer(),
        "day_name": Text(),
    }
)

order_item_refunds.to_sql(
    name="order_item_refunds",
    con=engine,
    if_exists="append",
    index=False,
    dtype={
        "order_item_refund_id": BigInteger(),
        "created_at": DateTime(),
        "order_item_id": BigInteger(),
        "order_id": BigInteger(),
        "refund_amount_usd": Numeric(10, 2),
        "year": Integer(),
        "month": Integer(),
        "day": Integer(),
        "day_name": Text(),
    }
)






```
